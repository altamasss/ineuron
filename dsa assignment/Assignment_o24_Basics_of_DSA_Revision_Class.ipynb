```python
def romanToInt(s):
    roman_values = {
        'I': 1,
        'V': 5,
        'X': 10,
        'L': 50,
        'C': 100,
        'D': 500,
        'M': 1000
    }

    result = 0

    for i in range(len(s) - 1):
        if roman_values[s[i]] < roman_values[s[i + 1]]:
            result -= roman_values[s[i]]
        else:
            result += roman_values[s[i]]

    result += roman_values[s[-1]]

    return result



print(romanToInt("III"))
print(romanToInt("LVIII"))
```

```python
def lengthOfLongestSubstring(s):
    start = 0
    end = 0
    max_length = 0
    visited = set()

    while end < len(s):
        if s[end] not in visited:
            visited.add(s[end])
            max_length = max(max_length, end - start + 1)
            end += 1
        else:
            visited.remove(s[start])
            start += 1

    return max_length



print(lengthOfLongestSubstring("abcabcbb"))
print(lengthOfLongestSubstring("bbbbb"))
print(lengthOfLongestSubstring("pwwkew"))
```

```python
def majorityElement(nums):
    majority = None
    count = 0

    for num in nums:
        if count == 0:
            majority = num
            count = 1
        elif num == majority:
            count += 1
        else:
            count -= 1

    return majority



print(majorityElement([3, 2, 3]))
print(majorityElement([2, 2, 1, 1, 1, 2, 2]))
```

```python
def groupAnagrams(strs):
    anagramMap = {}

    for s in strs:
        sorted_s = ''.join(sorted(s))
        if sorted_s not in anagramMap:
            anagramMap[sorted_s] = []
        anagramMap[sorted_s].append(s)

    return list(anagramMap.values())


print(groupAnagrams(["eat", "tea", "tan", "ate", "nat", "bat"]))


print(groupAnagrams(["", ""]))


print(groupAnagrams(["a"]))

```

```python
def nthUglyNumber(n):
    ugly = [0] * n
    ugly[0] = 1
    p2, p3, p5 = 0, 0, 0

    for i in range(1, n):
        next_ugly = min(2 * ugly[p2], 3 * ugly[p3], 5 * ugly[p5])
        ugly[i] = next_ugly

        if next_ugly == 2 * ugly[p2]:
            p2 += 1
        if next_ugly == 3 * ugly[p3]:
            p3 += 1
        if next_ugly == 5 * ugly[p5]:
            p5 += 1

    return ugly[n-1]

print(nthUglyNumber(10))
print(nthUglyNumber(1))
```

```python
from collections import defaultdict

def topKFrequent(words, k):
    word_freq = defaultdict(int)

    for word in words:
        word_freq[word] += 1

    sorted_words = sorted(word_freq.keys(), key=lambda x: (-word_freq[x], x))
    return sorted_words[:k]


words1 = ["i", "love", "leetcode", "i", "love", "coding"]
k1 = 2
print(topKFrequent(words1, k1))

words2 = ["the", "day", "is", "sunny", "the", "the", "the", "sunny", "is", "is"]
k2 = 4
print(topKFrequent(words2, k2))
```

```python
from collections import deque

def maxSlidingWindow(nums, k):
    window = deque()
    result = []
    for i in range(len(nums)):
        while window and window[0] <= i - k:
            window.popleft()
        while window and nums[window[-1]] <= nums[i]:
            window.pop()
        window.append(i)
        if i >= k - 1:
            result.append(nums[window[0]])
    return result
nums1 = [1, 3, -1, -3, 5, 3, 6, 7]
k1 = 3
print(maxSlidingWindow(nums1, k1))


nums2 = [1]
k2 = 1
print(maxSlidingWindow(nums2, k2))
```

```python
def findClosestElements(arr, k, x):
    left = 0
    right = len(arr) - 1
    while right - left >= k:
        if abs(arr[left] - x) <= abs(arr[right] - x):
            right -= 1
        else:
            left += 1
    return arr[left:right+1]
arr1 = [1, 2, 3, 4, 5]
k1 = 4
x1 = 3
print(findClosestElements(arr1, k1, x1))
arr2 = [1, 2, 3, 4, 5]
k2 = 4
x2 = -1
print(findClosestElements(arr2, k2, x2))
```
